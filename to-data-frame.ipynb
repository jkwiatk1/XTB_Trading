```python
import logging
import json
import xapi
import pandas as pd
```

```python
logging.basicConfig(level=logging.INFO)

with open("credentials.json", "r") as f:
    CREDENTIALS = json.load(f)
```

```python
my_columns = ['Ticker', 'Ask','Bid','Spread','Description','Ask/Bid Time']
final_dataframe = pd.DataFrame(columns=my_columns)
```

```python
x = await xapi.connect(**CREDENTIALS)
response = await x.socket.getAllSymbols()
```

```python
for item in response['returnData']:
    final_dataframe = final_dataframe.append(
        pd.Series(
            [
                item['symbol'],
                item['ask'],
                item['bid'],
                item['spreadRaw'],
                item['description'],
                item['time']
            ],
            index = my_columns
        ),
        ignore_index=True
    )
```

```python
final_dataframe
```

```python
def divide_chunks(lst, n):
    for i in range(0, len(lst), n):
        yield lst[i:i + n]
```

```python
symbol_groups = list(divide_chunks(final_dataframe['Ticker'],100))
```

```python
symbol_strings = []
for i in range(0,len(symbol_groups)):
    symbol_strings.append(','.join(symbol_groups[i]))
    print(symbol_strings[i])
```
